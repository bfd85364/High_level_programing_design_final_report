```python
year_by_df_ott_genre=pd.read_csv('2021-2025_df_ott_genre.csv', encoding='cp949')
# '시점', '구분별(1)', '구분별(2)' 컬럼을 기준으로 데이터를 그룹화합니다.
year_and_age_grouped_data = year_by_df_ott_genre.groupby(['시점', '구분별(1)', '구분별(2)']).sum(numeric_only=True)
display(year_and_age_grouped_data)
```

```python
import pandas as pd
import numpy as np

# Make a copy to avoid modifying the original dataframe
year_by_total_ott_genre_cleaned = year_by_total_ott_genre.copy()

# Convert columns with '-' to numeric, treating '-' as NaN
for col in ['영화 (%)', '예능/오락(개인방송 콘텐츠 포함) (%)', '드라마 (%)', '교양/다큐멘터리 (%)',
            '교육/학습 (%)', '애니메이션(만화) (%)', '스포츠 (%)', '뉴스 (%)',
            '공연(콘서트 연극 뮤지컬 등) (%)', '기타 (%)']:
    if col in year_by_total_ott_genre_cleaned.columns:
        # Replace '-' with NaN, then convert to numeric
        year_by_total_ott_genre_cleaned[col] = year_by_total_ott_genre_cleaned[col].replace('-', np.nan).astype(float)

# Reset index of year_and_age_grouped_data for merging
year_and_age_grouped_data_reset = year_and_age_grouped_data.reset_index()

# Identifier columns for merging
id_columns = ['시점', '구분별(1)', '구분별(2)']

# Identify common numeric columns for comparison (excluding id_columns)
# Filter for columns that are numeric in both dataframes (after cleaning)
common_numeric_cols_total = year_by_total_ott_genre_cleaned.select_dtypes(include=np.number).columns.drop('시점', errors='ignore')
common_numeric_cols_grouped = year_and_age_grouped_data_reset.select_dtypes(include=np.number).columns.drop('시점', errors='ignore')

comparison_columns = list(set(common_numeric_cols_total) & set(common_numeric_cols_grouped))

# Perform an outer merge to capture all possible rows from both dataframes
merged_for_comparison = pd.merge(
    year_by_total_ott_genre_cleaned,
    year_and_age_grouped_data_reset,
    on=id_columns,
    how='outer',
    suffixes=('_total', '_grouped')
)

# Initialize a dictionary to store discrepancies
discrepancies = {}

# Iterate through each common numeric column to find differences
for col in comparison_columns:
    col_total = f"{col}_total"
    col_grouped = f"{col}_grouped"

    # Check for three types of discrepancies:
    # 1. Values exist in both but are not close (for floats)
    # 2. Value exists in '_total' but is NaN in '_grouped' (missing in grouped data)
    # 3. Value exists in '_grouped' but is NaN in '_total' (missing in total data)

    # Mask for case 1: both non-NaN and numerically different
    mask_diff_values = (
        merged_for_comparison[col_total].notna() &
        merged_for_comparison[col_grouped].notna() &
        ~np.isclose(merged_for_comparison[col_total], merged_for_comparison[col_grouped], rtol=1e-05, atol=1e-08, equal_nan=False)
    )

    # Mask for case 2: missing in _grouped (value in _total, NaN in _grouped)
    mask_missing_grouped = (
        merged_for_comparison[col_total].notna() &
        merged_for_comparison[col_grouped].isna()
    )

    # Mask for case 3: missing in _total (NaN in _total, value in _grouped)
    mask_missing_total = (
        merged_for_comparison[col_total].isna() &
        merged_for_comparison[col_grouped].notna()
    )

    # Combine all discrepancy masks
    total_discrepancy_mask = mask_diff_values | mask_missing_grouped | mask_missing_total

    if total_discrepancy_mask.any():
        discrepancies[col] = merged_for_comparison[total_discrepancy_mask][id_columns + [col_total, col_grouped]]

# Display results
if discrepancies:
    print("`year_by_total_ott_genre`와 `year_and_age_grouped_data` 변수 간의 교차 검증 결과, 다음과 같은 불일치가 발견되었습니다:\n")
    for col, diff_df in discrepancies.items():
        print(f"컬럼: '{col}'에서 불일치\n")
        display(diff_df)
        print("\n" + "="*80 + "\n")
else:
    print("`year_by_total_ott_genre`와 `year_and_age_grouped_data` 변수 간의 공통 숫자 컬럼에서 중요한 불일치가 발견되지 않았습니다.")
```
