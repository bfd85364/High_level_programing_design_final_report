```python
df_total = pd.concat([df_2019, df_2020, df_2021, df_2022, df_2023, df_2024, df_2025], keys=[2019, 2020, 2021, 2022, 2023, 2024, 2025], axis=0)
df_total = df_total.reset_index(level=0)
df_total = df_total.rename(columns={'level_0': 'year'})
df_total['관객수'] = df_total['관객수'].astype(str).str.replace(',', '', regex=False).astype(int)

df_total_annual_by_type = df_total.groupby(['year', '상영타입'])['관객수'].sum().reset_index()
df_total_annual_by_type = df_total_annual_by_type.sort_values(by=['year', '관객수'], ascending=[True, False])
df_total_annual_by_type
```

```python
#관객수별 증가
import matplotlib.pyplot as plt
import seaborn as sns

df_total['상영타입'] = df_total['상영타입'].replace({'DOLBYCINEMA 더빙': 'DOLBYCINEMA dub'})

special_cinema_types = ['DOLBYCINEMA', 'VR', 'IMAX', '4D', 'ScreenX', 'DOLBYCINEMA 3D', 'DOLBYCINEMA dub']
df_special_cinema = df_total[df_total['상영타입'].isin(special_cinema_types)]

annual_audience_by_type = df_special_cinema.groupby(['year', '상영타입'])['관객수'].sum().reset_index()
annual_audience_by_type.fillna(0, inplace=True);

plt.figure(figsize=(12, 7))
sns.lineplot(data=annual_audience_by_type, x='year', y='관객수', hue='상영타입', marker='o')
plt.title('Annual Audience Trend by Special Cinema Type', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Audience', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(annual_audience_by_type['year'].unique())
plt.ticklabel_format(style='plain', axis='y') # Disable scientific notation for y-axis
plt.legend(title='Screening Type')
plt.tight_layout()
plt.show()
```

```python
df_movie_genre=pd.read_csv("movie_genre_final.csv")
df_movie_genre
```

```python
if pd.api.types.is_object_dtype(df_movie_genre['관객수_명']):
    df_movie_genre['관객수_명'] = df_movie_genre['관객수_명'].str.replace(',', '').astype(int)

idx = df_movie_genre.groupby('장르')['관객수_명'].idxmax()

max_audience_per_genre = df_movie_genre.loc[idx, ['장르', '연도', '관객수_명']]
max_audience_per_genre.sort_values(by='관객수_명', ascending=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

annual_special = df_total.groupby('year')['관객수'].sum().reset_index()
annual_special.columns = ['Year', 'Special_Cinema_Audience']

annual_anime = df_movie_genre[df_movie_genre['장르'] == '애니메이션'][['연도', '관객수_명']]
annual_anime.columns = ['Year', 'Anime_Audience']


df_corr = pd.merge(annual_special, annual_anime, on='Year')
correlation = df_corr['Special_Cinema_Audience'].corr(df_corr['Anime_Audience'])
print(f"Correlation between Special Cinema and Anime Audience: {correlation:.4f}")


fig, ax1 = plt.subplots(figsize=(12, 6))
color1 = 'tab:blue'
ax1.set_xlabel('Year')
ax1.set_ylabel('Special Cinema Audience', color=color1)
ax1.plot(df_corr['Year'], df_corr['Special_Cinema_Audience'], color=color1, marker='o', label='Special Cinema')
ax1.tick_params(axis='y', labelcolor=color1)

ax2 = ax1.twinx()
color2 = 'tab:red'
ax2.set_ylabel('Anime Audience', color=color2)
ax2.plot(df_corr['Year'], df_corr['Anime_Audience'], color=color2, marker='s', label='Anime')
ax2.tick_params(axis='y', labelcolor=color2)

plt.title('Trend Comparison: Special Cinema vs Anime Audience')
fig.tight_layout()
plt.show()
```
