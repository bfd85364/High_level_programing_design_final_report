```python
df_cinema_cs = pd.read_excel('phase_of_cinema_customer_covid19.xlsx')
df_cinema_cs.columns = ['연도', '원하는_영화상영', '극장위치', '상영시간', '스크린크기', '음향품질', '티켓구입_편의성', '상영관_청결', '특수상영관이용', '매점_및_주변상권', '직원응대', '극장_이벤트_굿즈', '좌석편의성', '티켓가격', '주차_교통', '극장안전_보안', '코로나19_방역', '주변인_선호', '부대시설']
df_cinema_cs.head()
```

```python
df_cinema_cs.dropna(subset=['연도'], inplace=False)
df_cinema_cs['연도'] = df_cinema_cs['연도'].astype('int64')
df_cinema_cs.sort_values(by='연도', inplace=False)
df_cinema_cs.head()
```

```python
year_list = df_cinema_cs.set_index('연도')
year_list.sort_values(by='연도', inplace=False)
```

```python
year_list = year_list.fillna(0)
year_list.sort_values(by='연도', inplace=False)
```

```python
special_audience = df_total.groupby('year')['관객수'].sum()

satisfaction_data = df_cinema_cs[['연도', '특수상영관이용']].copy()
satisfaction_data.columns = ['year', 'satisfaction']
satisfaction_data = satisfaction_data.dropna()

comparison_df = pd.DataFrame({'special_vol': special_audience}).reset_index()
comparison_df = pd.merge(comparison_df, satisfaction_data, on='year', how='inner')

correlation = comparison_df['special_vol'].corr(comparison_df['satisfaction'])
print(f"Correlation between Special Theater Audience Volume and Satisfaction: {correlation:.4f}")

fig, ax1 = plt.subplots(figsize=(10, 6))
ax1.set_xlabel('Year')
ax1.set_ylabel('Special Theater Audience Volume', color='tab:blue')
ax1.plot(comparison_df['year'], comparison_df['special_vol'], marker='o', color='tab:blue', label='Audience Volume')
ax1.tick_params(axis='y', labelcolor='tab:blue')

ax2 = ax1.twinx()
ax2.set_ylabel('Satisfaction Score', color='tab:red')
ax2.plot(comparison_df['year'], comparison_df['satisfaction'], marker='s', color='tab:red', label='Satisfaction')
ax2.tick_params(axis='y', labelcolor='tab:red')

plt.title('Special Theater: Audience Volume vs Satisfaction')
fig.tight_layout()
plt.show()
```

데이터 전처리 및 병합: df_total에서 연도별 특수관 총 관객수(special_vol)를 집계하고, 설문 데이터(df_cinema_cs)에서 특수관 이용 만족도 데이터를 추출하여 '연도(year)'를 기준으로 하나의 표(comparison_df)로 합칩니다.
상관관계 계산: corr() 함수를 사용하여 관객 수와 만족도 사이의 상관계수를 계산합니다. 실행 결과 -0.9150이 나왔는데, 이는 관객 수가 늘어날수록 만족도는 낮아지는 강한 음의 상관관계가 있음을 의미합니다.
