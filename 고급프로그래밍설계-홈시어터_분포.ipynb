```python
import pandas as pd

file_names = [
    '2019_home_watcher.csv',
    '2020_home_watcher.csv',
    '2021_home_watcher.csv',
    '2022_home_watcher.csv',
    '2023_home_watcher.csv',
    '2024_home_watcher.csv'
]

all_dfs = []

for file in file_names:
    year = int(file.split('_')[0])

    df_home_watcher = pd.read_csv(file, encoding='cp949', header=None)

    df_home_watcher[0] = df_home_watcher[0].astype(str).str.replace(r'\s+', '', regex=True)
    df_home_watcher[1] = df_home_watcher[1].astype(str).str.replace(r'\s+', '', regex=True)

    df_filtered = df_home_watcher[df_home_watcher[0] == '가구주연령'].copy()

    df_result = pd.DataFrame({
        '연도': year,
        '연령대': df_filtered[1],
        '스마트TV_보급률(%)': pd.to_numeric(df_filtered[8], errors='coerce')
    })

    all_dfs.append(df_result)

df_final = pd.concat(all_dfs, ignore_index=True)
df_final
```

```python
#pyplot 방식 -> 자동

#객체 지향 API 방식 (Figure, Axes 객체 중심의 유연한 플롯 수성) -> pyplot모듈에서 plt.subplots()->Figure와 Axes를 한번에 만들어주는 편의
#함수, fig.add_subplot() -> Figure를 먼저 만들고 Axex 하나씩 추가
# Objectfig, ax 객체를 만든 뒤,ax.plot(), ax.bar()처럼 그래프 영역에 직접 그림


#현재의 방식 OOP 방식이 더 적합한가?

#예시) -> fig.add_subplot(211) -> 211이 아니라, (2,1,1) 의미함. 즉, 2행 1열의 subplot중 1번째(첫번째) 위치를 의미
import plotly.express as px

# Ensure '연도' is treated as a categorical variable for plotting, but also sort it numerically
df_final_sorted = df_final.sort_values(by='연도').copy()

fig = px.line(
    df_final_sorted,
    x='연도',
    y='스마트TV_보급률(%)',
    color='연령대',
    title='연령대별 스마트 TV 보급률 변화',
    labels={
        '연도': '연도',
        '스마트TV_보급률(%)': '스마트 TV 보급률 (%)',
        '연령대': '연령대'
    },
    markers=True, # Add markers for each data point
    hover_data={'연령대': True, '스마트TV_보급률(%)': ':.1f'}
)

fig.update_layout(hovermode="x unified")
fig.show()
```
