```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Re-configure matplotlib for the new font
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic')
plt.rcParams['axes.unicode_minus'] = False
```

```python
ds = pd.read_csv('/content/배속 이용.csv')
ds.head()
```

```python
plt.rc('font', family='NanumBarunGothic')
plt.rcParams['axes.unicode_minus'] = False
```

```python
ds_ages = ds[ds['연령대'] != '전체']
```

```python
plt.figure(figsize=(10, 6))
colors = sns.color_palette("Blues_d", len(ds_ages))
```

```python
bars = plt.bar(ds_ages['연령대'], ds_ages['배속재생이용률(%)'], color=colors, width=0.6, edgecolor='gray', linewidth=0.5)

plt.title('연령대별 배속재생 이용률 (%)', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('연령대', fontsize=12, labelpad=10)
plt.ylabel('이용률 (%)', fontsize=12, labelpad=10)
plt.ylim(0, 30)
plt.grid(axis='y', linestyle=':', alpha=0.6)

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, height + 0.5, f'{height}%',
             ha='center', va='bottom', fontsize=11, fontweight='bold')

plt.tight_layout()
plt.show()
```

### Install Nanum Fonts
To properly display Korean characters, we need to install the Nanum fonts and configure matplotlib to use them.

After installing the fonts, we need to set them as the default font in matplotlib. This often requires restarting the runtime for the changes to take full effect. Once the runtime is restarted, you can re-run the plotting code.
