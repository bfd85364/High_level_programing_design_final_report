```python
from pandas import DataFrame
import matplotlib.pyplot as plt
import pandas as pd
import sys
import numpy as np
```

```python
df_cinema_cs = pd.read_excel('phase_of_cinema_customer_covid19.xlsx')
df_cinema_cs.columns = ['연도', '원하는_영화상영', '극장위치', '상영시간', '스크린크기', '음향품질', '티켓구입_편의성', '상영관_청결', '특수상영관이용', '매점_및_주변상권', '직원응대', '극장_이벤트_굿즈', '좌석편의성', '티켓가격', '주차_교통', '극장안전_보안', '코로나19_방역', '주변인_선호', '부대시설']
df_cinema_cs.head()
```

```python
#연도 칼럼의 결측값 제거 -> dropna 사용하기
df_cinema_cs.dropna(subset=['연도'], inplace=False)
df_cinema_cs['연도'] = df_cinema_cs['연도'].astype('int64')
df_cinema_cs.sort_values(by='연도', inplace=False)
df_cinema_cs.head()
```

```python
year_list = df_cinema_cs.set_index('연도')
year_list.sort_values(by='연도', inplace=False)
```

```python
year_list = year_list.fillna(0)
year_list.sort_values(by='연도', inplace=False)
```

```python
columns_to_plot = ['원하는_영화상영','상영시간', '스크린크기', '음향품질', '특수상영관이용']
for col in columns_to_plot:
    year_list[col] = pd.to_numeric(year_list[col], errors='coerce')

col_labels = ['prefer_movie', 'running_time', 'screen_size', 'sound_quality', 'special_cinema']

plt.figure(figsize=(12, 7))

for i, col in enumerate(columns_to_plot):
    plt.plot(year_list.index, year_list[col], label=col_labels[i])

plt.title('year of customer satisfaction about [movie experiance]')
plt.xlabel('year')
plt.ylabel('Customer Satisfaction Score')

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.xticks(rotation=45, ha='right')

plt.tight_layout()

plt.show()
```

```python
#도출된 결과: 지난 극장 조사의 판데믹 전중후중  후기에 포함되는 2022년 이후의 예상 인사이트에서 특수상영관의 이용율이 증가하는 것을 관측한 것과 별개로 실제로 만족도 조사에서는 특수 상영관 이용에 대한 만족도가 감소하는 것을 확인함
#따라서, 이용율이 증가하는데 비해, 만족도가 감소하는 것에 대해서는 해석이 필요
```

```python
#결과에 대한 탐색 접근법
# - 연도별 영화 장르 편중
#-  장르별 특수관 이용
#- 관객수 비교: 연도별 총 누적 관객수를 비교하여 상위와 하위를 나누고 상위에 해당하는 그룹과 하위에 해당하는 그룹간의 차이점을 찾아서 해석에 포함될 요인을 찾는다
```

```python
columns_to_plot = ['매점_및_주변상권','부대시설', '주차_교통', '극장위치']
for col in columns_to_plot:
    year_list[col] = pd.to_numeric(year_list[col], errors='coerce')
    # 플로팅 전에 결측치를 0으로 명시적으로 채웁니다.
    year_list[col] = year_list[col].fillna(0)
    #[주의!!]: 2021 ~ 2025년 사이의 주차_교통(parking_transport) 칼럼 그래프와, 부대시설(infra_facility) 칼럼 그래프는
    #          0으로 감소하는 것이 아니라, 결측치를 0으로 처리한것이다!

col_labels = ['store&market', 'infra_facility', 'parking_transport', 'cinema_location']

plt.figure(figsize=(12, 7))

for i, col in enumerate(columns_to_plot):
    plt.plot(year_list.index, year_list[col], label=col_labels[i], marker='o')
plt.title('year of customer satisfaction about [cinema Infra system]')
plt.xlabel('year')
plt.ylabel('Customer Satisfaction Score')

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.xticks(rotation=45, ha='right')

plt.tight_layout()

plt.show()
```

```python
df_ticket_price = pd.read_csv('multiplex_price_trends.csv')
df_ticket_price
```

```python
year_ticket_price = df_ticket_price.groupby('인상 시점').max()
year_ticket_price.sort_values(by='인상 시점', ascending=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df_ticket_price['year'] = pd.to_datetime(df_ticket_price['인상 시점']).dt.year
actual_price_yearly = df_ticket_price.groupby('year')['주말 가격'].apply(lambda x: x.str.replace(',', '').astype(float).mean()).reset_index()
actual_price_yearly.columns = ['Year', 'Actual_Price']

satisfaction_data = year_list[['티켓가격']].copy()
satisfaction_data.index.name = 'Year'
satisfaction_data = satisfaction_data.reset_index()
satisfaction_data.columns = ['Year', 'Satisfaction_Score']

merged_df = pd.merge(actual_price_yearly, satisfaction_data, on='Year', how='inner')
correlation = merged_df['Actual_Price'].corr(merged_df['Satisfaction_Score'])
print(f"Correlation between Actual Price and Satisfaction: {correlation:.4f}")

fig, ax1 = plt.subplots(figsize=(10, 6))
color1 = 'tab:red'
ax1.set_xlabel('Year')
ax1.set_ylabel('Actual Ticket Price (KRW)', color=color1)
ax1.plot(merged_df['Year'], merged_df['Actual_Price'], color=color1, marker='o', label='Actual Price')
ax1.tick_params(axis='y', labelcolor=color1)

ax2 = ax1.twinx()
color2 = 'tab:blue'
ax2.set_ylabel('Satisfaction Score', color=color2)
ax2.plot(merged_df['Year'], merged_df['Satisfaction_Score'], color=color2, marker='s', label='Satisfaction')
ax2.tick_params(axis='y', labelcolor=color2)

plt.title('Comparison of Actual Ticket Price and Customer Satisfaction')
fig.tight_layout()
plt.show()
```

분석 결과, 실제 티켓 가격과 티켓 가격 만족도 사이에는 -0.8973이라는 매우 강한 음의 상관관계가 나타났습니다.

이는 통계적으로 유의미한 상관관계이긴 하지만 실제 티켓 가격이 상승할수록 사용자의 만족도는 뚜렷하게 감소함을 의미합니다.
